```python
from unidecode import unidecode
output = []
main_text=[]
for page in doc:
    output += page.get_text("blocks")
previous_block_id = 0 # Set a variable to mark the block id
for block in output:
    if block[6] == 0: # We only take the text
        if previous_block_id != block[5]:# Compare the block number
            print("\n")
            plain_text = unidecode(block[4])
            main_text.append(plain_text)
            print(plain_text)
```

```python
# import spacy
import spacy
from spacy.scorer import Scorer
#g=final_list[1]
# load spacy model
nlp = spacy.load('en_core_web_sm')
# load data
#CARDINAL,FAC, LANGUAGE, LAW,NORP,ORDINAL,PERCENT,PROD,QUAN,WORK OF ART
spacy_dict={'CARDINAL':[],
            'DATE':[],
            'EVENT':[],
             'FAC':[],
             'GPE':[],
             'LANGUAGE':[],
             'LAW':[],
             'LOC':[],
             'MONEY':[],
             'NORP':[],
             'ORDINAL':[],
             'ORG':[],
             'PERCENT':[],
             'PERSON':[],
             'PRODUCT':[],
             'QUANTITY':[],
             'TIME':[],
             'WORK_OF_ART':[]          
}
for sentence in final_list:
    doc = nlp(sentence)
    # print entities
    temp_dict={'CARDINAL':[],
            'DATE':[],
            'EVENT':[],
             'FAC':[],
             'GPE':[],
             'LANGUAGE':[],
             'LAW':[],
             'LOC':[],
             'MONEY':[],
             'NORP':[],
             'ORDINAL':[],
             'ORG':[],
             'PERCENT':[],
             'PERSON':[],
             'PRODUCT':[],
             'QUANTITY':[],
             'TIME':[],
             'WORK_OF_ART':[]}
    li=[]
    for ent in doc.ents:
        li.append(ent.label_)
        temp_dict[ent.label_].append(ent.text)
    for k in temp_dict:
        if len(temp_dict[k]) != 0:
            spacy_dict[k].append(temp_dict[k])
    for j in temp_dict:
        if j not in li:
            spacy_dict[j].append('-')
```

```python
# Step One: Import nltk and download necessary packages
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
spacy1_dict={ 'ORGANIZATION':[],
                 'PERSON':[],
                 'LOCATION':[],
                 'DATE':[],
                 'TIME':[],
                 'MONEY':[],
                 'PERCENT':[],
                 'FACILITY':[],
                  'GPE':[] }
for sentence in final_list:

        # print entitie
    temp_dict={ 'ORGANIZATION':[],
                 'PERSON':[],
                 'LOCATION':[],
                 'DATE':[],
                 'TIME':[],
                 'MONEY':[],
                 'PERCENT':[],
                 'FACILITY':[],
                  'GPE':[]   }
    li=[]
    for sent in nltk.sent_tokenize(sentence):
        for chunk in nltk.ne_chunk(nltk.pos_tag(nltk.word_tokenize(sent))):
            if hasattr(chunk, 'label'):
                li.append(chunk.label())
                temp_dict[chunk.label()].append(chunk[0][0])
    for k in temp_dict:
        if len(temp_dict[k]) != 0:
             spacy1_dict[k].append(temp_dict[k])
    for j in temp_dict:
         if j not in li:
            spacy1_dict[j].append('-')
```
